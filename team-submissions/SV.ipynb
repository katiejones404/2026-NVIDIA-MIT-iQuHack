```python
# Self-Validation - automating hand calculations (up to small N)

def C_k(s, k):
    N = len(s)
    sum = 0
    for i in range (N-k):
        next = s[i]*s[i+k]
        sum = sum + next
    return sum

def E(s):
    N = len(s)
    sum = 0
    for k in range (1,N):
        next = C_k(s, k)**2
        sum = sum + next
    return sum

from itertools import product

def sequence(N):
    return [list(seq) for seq in product([1, -1], repeat=N)]

def calculate(N):
    Elist = []
    seq = sequence(N)
    for j in range(len(seq)):
        sj = seq[j]
        Ej = E(sj)
        Elist.append(Ej)
    minE = min(Elist)
    mini = [index for index, value in enumerate(Elist) if value == minE]
    print(f"Optimal Energy: {minE}")
    print(f"Number of Optimal Sequences: {len(mini)}")
    for m in mini:
        print(f"Optimal Sequence: {seq[m]}")
```

```python
calculate(10)
```
